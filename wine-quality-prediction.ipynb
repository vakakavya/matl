```r
library(caTools)
set.seed(100)

wine_data$response<-ifelse(wine_data$quality >= 6, 1, 0)
split<-sample.split(wine_data, SplitRatio=0.65)
train<-subset(wine_data, split==T)
test<-subset(wine_data, split==F)

head(train)
tail(train)

head(test)
tail(test)

ggplot(train, aes(x=response, fill=factor(quality)))+geom_bar()
```

```r
p<-density(wine_data$quality)
plot(p)
```

```r
#checking significance values
#sulphates, alcohol, citric.acid, fixed.acidity
t.test(wine_data$citric.acid, wine_data$quality)
boxplot(wine_data$citric.acid, wine_data$quality)


t.test(wine_data$sulphates, wine_data$quality)
boxplot(wine_data$sulphates, wine_data$quality)

t.test(wine_data$alcohol, wine_data$quality)
boxplot(wine_data$alcohol, wine_data$quality)


t.test(wine_data$fixed.acidity, wine_data$quality)
boxplot(wine_data$fixed.acidity, wine_data$quality)
```

```r
#sulphates, alcohol, citric.acid, fixed.acidity
#Logistic Model
library(caret)
options(scipen=0999)

log_model<-glm(response ~., data=train)

predicted_model <- predict(log_model, test, type="response") 

#Threshold value for confusion matrix is 0.5
#confusionMatrix(test$response, predicted_model>0.5)

str(test$response)
str(predicted_model)
```

```r
library(caTools)
set.seed(100)

wine_data$response<-ifelse(wine_data$quality >= 6, 1, 0)
split<-sample.split(wine_data, SplitRatio=0.65)
train<-subset(wine_data, split==T)
test<-subset(wine_data, split==F)

head(train)
tail(train)

head(test)
tail(test)



```

```r
#sulphates, alcohol, citric.acid, fixed.acidity
#Logistic Model
library(caret)
options(scipen=0999)

log_model<-glm(response ~., data=train)

predicted_model <- predict(log_model, test, type="response") 
pred<-as.integer(predicted_model)
#pred
summary(pred)
summary(log_model)
#Threshold value for confusion matrix is 0.5
#confusion matrix
table(test$response, pred>0.5)

str(test$response)
str(pred)

#accuracy
(287+85)/(287+85+243)

#to improve the accuracy
log_new_model<-glm(response ~.-volatile.acidity-chlorides-total.sulfur.dioxide-density-pH-fixed.acidity, data = train)
new_predicted_model<-predict(log_new_model, test, type='response')
table(test$response, as.integer(new_predicted_model)>0.5)
```

```r
Submission<-data.frame(Quality=new_predicted_model)
Submission

#Create submission file
write.csv(Submission, file='Predictions.csv', row.names=F)
```
